```python
import pandas as pd
```

```python
data = pd.read_excel(r"C:\Users\Acer-PC\Documents\Panda\Customer Call List.xlsx")
data
```

```python
# Removing Duplicates
data.drop_duplicates()
```

```python
# Removing Columns
data.drop(columns = "Not_Useful_Column")
```

```python
# Strip left of the string
# data["Last_Name"] = data["Last_Name"].str.lstrip("...")

# Strip right of the string
# data["Last_Name"] = data["Last_Name"].str.rstrip("_")

# Strip remove on both side
data["Last_Name"] = data["Last_Name"].str.strip("123._/")
data
```

```python
# Cleaning phone number

#Remove value except lower and upper case letters and 0-9
data["Phone_Number"] = data["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
data
```

```python
# Adding slash or dash between the number and return them
# Convert data type to string
data["Phone_Number"] = data["Phone_Number"].apply(lambda x: str(x))
data
```

```python
# Add the dash or slash between the phone number
data["Phone_Number"] = data["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
data
```

```python
# Replacing nan-- and Na-- value
data["Phone_Number"] = data["Phone_Number"].str.replace('nan--', '')
data["Phone_Number"] = data["Phone_Number"].str.replace('Na--', '')
data
```

```python
# Split Address

data[["Street_Address", "State", "Zip_Code"]] = data["Address"].str.split(',', 2, expand = True)
data
```

```python
# Replacing Yes and No value to Y and N

data["Paying Customer"] = data["Paying Customer"].str.replace('Yes', 'Y')

data["Paying Customer"] = data["Paying Customer"].str.replace('No', 'N')
data
```

```python
data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace('Yes', 'Y')

data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace('No', 'N')
data
```

```python
# Replacing N/a and Nan value 

data = data.replace('N/a', '')
data = data.replace('NaN', '')

data = data.fillna('')
data
```

```python
# Filtering Down Rows to Keep CustomerID that need to be contacted

for x in data.index:
    if data.loc[x, "Do_Not_Contact"] == 'Y':
        data.drop(x, inplace = True)

data
```

```python
# Fill in blank with 'N'
data["Do_Not_Contact"] = data["Do_Not_Contact"].replace('', 'N')
data
```

```python
# Filtering Down Rows to remove customer with no phone number

for x in data.index:
    if data.loc[x, "Phone_Number"] == '':
        data.drop(x, inplace = True)

data

# Alternative way to remove null value from dataset
# data = data.dropna(subset = "Phone_Number"), inplace = True)
```

```python
# Reorder the Index

data = data.reset_index(drop = True)
data
```
